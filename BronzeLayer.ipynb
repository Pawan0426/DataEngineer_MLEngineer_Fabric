```python
import requests
import json
```

```python
url = 'https://newsapi.org/v2/top-headlines?country=us&apiKey=4e71ed796f8543af933f376878c5ab9d'
```

```python
response = requests.get(url)
#display(response)
```

```python
if response.status_code == 200:
    data = response.json()
    data = data['articles']
    file_path = f'/lakehouse/default/Files/EnNews.json'
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)
else:
    print('Fail to fetch data, status code: ', response.status_code)
```
