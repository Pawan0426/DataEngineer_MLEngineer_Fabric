```python
from pyspark.sql.functions import col
from pyspark.sql.types import TimestampType
```

```python
df = spark.read.option('multiline', 'true').json(f'Files/EnNews.json')
display(df)
```

```python
df = df.select('author',
                'content',
                'description',
                'publishedAt',
                col('source.name').alias('name'),
                'title',
                'url')
```

```python
df.write.mode('append').saveAsTable('SilverNews')
```
